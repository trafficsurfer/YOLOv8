```python
from ultralytics import YOLO
from PIL import Image
```

```python
model_path = r'runs/detect/yolov8n_ru_dataset_trans/weights/best.pt'
```

### Картинки

```python
model = YOLO(model_path)
```

```python
#Картинка из интернета. Россия
results = model('results/example_image_1.jpg')

for r in results:
    im_array = r.plot()
    im = Image.fromarray(im_array[..., ::-1])
    im.save('results/example_image_1_predict_train_model.jpg') 
```

```python
#Картинка из теста. Не Россия
results = model('results/example_image_2.jpg')

for r in results:
    im_array = r.plot()
    im = Image.fromarray(im_array[..., ::-1])
    im.save('results/example_image_2_predict_train_model.jpg')
```

### Видео по API

```python
url = "http://127.0.0.1:8000/"

response = requests.get(url)
print(response.text)
```

```python
add_path = 'runs/detect/predict/'

url = "http://127.0.0.1:8000/uploadfile/"

path_small = "example_video_small.mp4"

path_big = 'example_video_big.mp4'
```

```python
%%time
with open(path_small, "rb") as video_file:
    files = {"file": (video_file.name, video_file, "video/mp4")}
    response = requests.post(url, files=files)

print(response.text)
```

```python
#Сохранение принятого файла
with open(f"results/{path_small}", "wb") as output_file:
    output_file.write(response.content)
```
